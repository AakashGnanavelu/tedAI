```python
import pandas as pd
import os 

df = pd.read_csv(os.path.join('data', 'data(portion).csv'), on_bad_lines = 'skip', header = None)
df.columns = ['URL', 'Title', 'Speaker', 'Date', 'Views', 'Likes', 'Topics', 'Description', 'Transcript']
```

```python
df.fillna(" ")
```

```python
text_col = []

for _, row in df.iterrows():
    prompt = "Below is an instruction that describes a task. Write a response that appropiately compeltes the request. \n"
    instruction = f"Create a speech using these topics {str(row['Topics'])}, with the title {str(row['Title'])}"
    response = str(row['Transcript'])

    text = prompt + '###Instruction: \n' + instruction + '\n### Response: \n' + response

    text_col.append(text)

df.loc[:, 'text'] = text_col
    
```

```python
df.head()
```

```python
df.to_csv('train.csv', index = False)
```
